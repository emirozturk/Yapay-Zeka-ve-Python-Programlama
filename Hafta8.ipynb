```python
#ax^2+bx+c=0 denkleminde a,b,c için disktriminant değerini bulan uygulama
#delta = b^2-4ac
#(-b-kök(delta))/2a 
#(-b+kök(delta))/2a
a=2
b=3
c=1
#2x^2+3x+1=0
delta = b*b-4*a*c
if delta < 0:
    print("Reel kök yok")
elif delta == 0:
    print((-b-math.sqrt(delta))/(2*a))
else:
    x1 = (-b-math.sqrt(delta))/(2*a)
    x2 = (-b+math.sqrt(delta))/(2*a)
    print(x1,x2)
```

```python
def f(x):
    return round(math.cos(math.radians(x)),2)
def f2(x):
    return round(math.sin(math.radians(x)),2)
def f3(x):
    return round(math.cos(math.radians(2*x)),2)
def f4(x):
    return round(math.sin(math.radians(2*x)),2)

liste = [i for i in range(0,720)]
cikti = [f(x) for x in liste]
cikti2 = [f2(x) for x in liste]
cikti3 = [f3(x) for x in liste]
cikti4 = [f4(x) for x in liste]

print(liste)
print(cikti)
print(cikti2)
```

```python
import matplotlib.pyplot as plt
plt.plot(liste,cikti,)
plt.plot(liste,cikti2,)
plt.plot(liste,cikti3,)
plt.plot(liste,cikti4,'x')
plt.legend(["cosx","sinx","cos2x","sin2x"],loc="upper right")
```
